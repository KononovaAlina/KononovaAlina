Исходя из приведенного (выше) графика, можно заключить, что среди "фаст-фуд" и "кулинария" харатерно сетевое распределение. Получается типам заведений "фаст-фуд" и "кулинария" характерней быть сетью, чем для "столовая", "буфет", "бар". Можно предположить, что для заведений типов "фаст-фуд" и "кулинария" выгодней прибегать к сетевом использованию с точки зрения будуших возможностей по масштабированию, а также за счет популярности данных типов заведений у потребителей. 

```python
#Код ревьюера, так можно не мержить
chain_by_rest_type = df.groupby('object_type').apply(lambda x: pd.Series(dict(
    chained=(x.chain == 'сетевое').sum(),
    not_chained=(x.chain == 'несетевое').sum(),
    total =x.id.count()
)))


chain_by_rest_type = chain_by_rest_type.reset_index()
chain_by_rest_type['chained_share'] = chain_by_rest_type['chained']/chain_by_rest_type['total']*100
chain_by_rest_type['not_chained_share'] = chain_by_rest_type['not_chained']/chain_by_rest_type['total']*100

chain_by_rest_type.sort_values(by='chained_share', ascending=False)
```

Исходя из проиведенных расчетов, можно заключить, что сетевая модель популярна среди "кафе", "фаст-фуд" и  "ресторан". Однако следует отметить, что сетевых "столовая" и "буфет" почти не имеется.

### Рассмотрим, что характерно для сетевых заведений.

```python
#Тут не очень понятно зачем группировать по числу мест, ведь зведения из одной сети с разным числом мест всё ещё в одной сети
cnt_seats_id_type  = df.groupby(['object_name', 'chain']) \
                        .agg({'id':'count', 'number_of_seats' : lambda x : round(x.mean())}).reset_index().sort_values(by='id', ascending=False)
cnt_seats_id_type.columns = ['object_name','chain', 'number_of_seats', 'count']
cnt_seats_id_type
```

```python
#Отсюда желательно  взять только сетевые, выкинуть сети с одним заведеним, чтобы они не смещали статистики для числа заведений 

cnt_seats_chain = cnt_seats_id_type[(cnt_seats_id_type['chain'] == 'сетевое') &
                                     (cnt_seats_id_type['count'] > 1)]


fig = px.scatter(cnt_seats_chain, x="count", y="number_of_seats", color="object_name",
                title = 'Сетевые: зависимость количества мест от количества заведений по каждому типу', hover_data=['object_name'])

# Медиана или среднее для числа мест
fig.add_vline(cnt_seats_chain['count'].mean())
fig.add_hline(cnt_seats_chain['number_of_seats'].mean())
fig.show()
```

```python
cnt_seats_id_type  = df.groupby(['object_name', 'chain', 'number_of_seats']) \
                        .agg({'id':'count'}).reset_index().sort_values(by='id', ascending=False)
cnt_seats_id_type.columns = ['object_name','chain', 'number_of_seats', 'count']
cnt_seats_id_type

top_10_object_name = cnt_seats_id_type.query('chain=="сетевое"').head(5)
top_10_object_name
```

## Выделим в отдельный столбец информацию об улице из столбца address.

Разобьем адрес по запятой.

```python
df['split_address'] = df['address'].str.split(', ')
```

Выделим улицу.

```python
def get_street(row):
    x = row['split_address'][1]
    return x
df['street'] = df.apply(get_street, axis=1)
```

Найдем 'дом' и 'город' в улицах.

```python
def get_adress_spec(row):
    house = re.search('дом$|^дом', row['street'])
    town = re.search('город$|^город', row['street'])
    if house == None:
        if town == None:
            return 'ок'
        else:
            return 'город'
    else:
        return 'дом'
df['house_or_town'] = df.apply(get_adress_spec, axis=1)
```

Укажем улицу, определим нужный индекс.

```python
def get_street_name(row):
    if row['house_or_town'] == 'дом':
        x = row['split_address'][0]
    else:
        if row['house_or_town'] == 'город':
            x = row['split_address'][2]
        else:
            x = row['split_address'][1]
    return x
df['street'] = df.apply(get_street_name, axis=1)
```

```python
df = df.drop(['house_or_town', 'split_address'], axis=1)
df.head()
```
